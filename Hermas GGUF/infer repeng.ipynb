```python
import base64, io, numpy as np

def npy_file_to_control_vector_b64(path: str) -> str:
    arr = np.load(path, allow_pickle=True)       # muat isi file .npy
    buf = io.BytesIO()
    np.save(buf, arr)                             # simpan ulang ke format .npy di memori
    return base64.b64encode(buf.getvalue()).decode("ascii")  # ‚Üê no newline
```

```python
import requests
import pprint

URL = "https://farhanaryawicaksono--lcpp-hermes2pro-handle-req.modal.run/"

b64_data = npy_file_to_control_vector_b64("kekbot.npy")  
payload = {
    "prompt": "Who are you?",
    "system_prompt": "You are a helpful assistant.",
    "gen_config": {"max_tokens": 512, "temperature": 0},
    "control": {
        "control_vector_npy_b64": b64_data,
        "control_strength": 0.4  # alias ke 'strength'
    }
}

r = requests.post(URL, json=payload, timeout=180)
r.raise_for_status()
pprint.pprint(r.json()["res"])
```
